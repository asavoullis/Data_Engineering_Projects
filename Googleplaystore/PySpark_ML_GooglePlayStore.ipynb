### Why Use PySpark?

- **Cluster Computing**: It can run on clusters of machines, enabling horizontal scalability. You can easily scale up or down based on your computing needs.

2. Speed

- **In-Memory Processing**: PySpark performs much of its data processing in-memory, which can significantly speed up data transformations and analysis.

- **Optimized Execution Plans**: Spark's Catalyst optimizer generates efficient execution plans, reducing query processing time.

3. Ease of Use

- **Pythonic API**: PySpark provides a Pythonic API, making it accessible to Python developers who may not have experience with Scala or Java, which are used in the native Spark API.

- **Integration**: It seamlessly integrates with Python libraries like Pandas, NumPy, and Matplotlib, allowing you to combine the power of Spark with familiar Python tools.

4. Rich Ecosystem

- **Machine Learning**: PySpark includes libraries like MLlib for machine learning tasks, enabling you to build and deploy large-scale machine learning models.

- **Streaming**: It supports real-time data processing through Spark Streaming, which is vital for applications like monitoring, fraud detection, and IoT data processing.

5. Data Source Agnosticism

- **Data Sources**: PySpark can read from various data sources, including HDFS, Apache Hive, Apache HBase, and many others, making it versatile for handling diverse data formats.

- **Connectivity**: It supports connecting to databases and external data sources, allowing you to work with structured and unstructured data from different origins.

6. Fault Tolerance

- **Resilience**: Spark's built-in fault tolerance ensures that if a node fails during processing, the task can be rerun on another node without data loss or disruption.

7. Community and Support

- **Large Community**: Apache Spark and PySpark have a thriving open-source community with extensive documentation, forums, and resources available online.

- **Active Development**: Regular updates and improvements are made to PySpark, ensuring that it stays relevant and up-to-date.

Apache Spark and Apache Hadoop are two separate but closely related open-source big data technologies. They are often used together as part of a modern big data stack. Here's an overview of their relationship:

1. **Apache Hadoop**:

   - **Hadoop Distributed File System (HDFS)**: Hadoop's primary storage component, HDFS, is a distributed file system designed to store and manage vast amounts of data across a cluster of commodity hardware. It provides fault tolerance, data replication, and distributed storage.

   - **MapReduce**: Hadoop introduced the MapReduce programming model, which is a framework for processing and generating large datasets in parallel across a distributed cluster. MapReduce is particularly well-suited for batch processing tasks.

   - **YARN (Yet Another Resource Negotiator)**: YARN is Hadoop's resource management and job scheduling component. It manages resources in a Hadoop cluster and allows multiple data processing frameworks to run on the same cluster efficiently.

2. **Apache Spark**:

   - **In-Memory Data Processing**: Spark is an open-source, distributed computing system that was designed to overcome some limitations of Hadoop MapReduce. It can perform in-memory data processing, which often results in significantly faster data processing compared to Hadoop's disk-based processing.

   - **Unified Data Processing**: Spark provides a unified platform for various data processing tasks, including batch processing, real-time stream processing, interactive queries, and machine learning. This versatility makes it more than just a batch processing framework.

   - **Ease of Use**: Spark offers user-friendly APIs in multiple programming languages, including Scala, Java, Python, and R. This accessibility has contributed to its popularity among data scientists and engineers.

3. **Relationship**:

   - **Integration**: Spark can be run on a Hadoop cluster, and it can read data directly from HDFS. This means that you can use Hadoop's HDFS as a storage layer and Spark as a processing engine, allowing you to leverage both technologies' strengths.

   - **Complementary**: While Hadoop and Spark can work together, they are also complementary. Spark can be seen as an improvement over Hadoop's MapReduce for certain use cases due to its in-memory processing and broader range of processing capabilities.

   - **Coexistence**: Many organizations use both Hadoop and Spark in their big data stack. Hadoop is often used for storage and batch processing, while Spark is used for interactive queries, real-time processing, and machine learning tasks.

In summary, Apache Spark and Apache Hadoop are part of the larger big data ecosystem. They can be used together or independently, depending on the specific needs of a data processing pipeline. Their relationship is one of coexistence and complementarity, with Spark often providing enhanced processing capabilities on top of Hadoop's distributed storage and resource management infrastructure.

```python
def plot_per_column_distribution(df: DataFrame, n_graph_shown: int, n_graph_per_row: int):
    # Select columns with between 1 and 50 unique values
    selected_cols = [col_name for col_name in df.columns if 1 < df.select(col_name).distinct().count() < 50]
    
    n_col = len(selected_cols)
    n_graph_row = (n_col + n_graph_per_row - 1) // n_graph_per_row
    
    plt.figure(figsize=(6 * n_graph_per_row, 8 * n_graph_row), dpi=80, facecolor='w', edgecolor='k')
    
    for i, col_name in enumerate(selected_cols[:n_graph_shown]):
        plt.subplot(n_graph_row, n_graph_per_row, i + 1)
        column_data = df.select(col_name).toPandas()
        
        if column_data[col_name].dtype == 'object':
            # For categorical columns
            sns.countplot(x=col_name, data=column_data)
        else:
            # For numeric columns
            sns.histplot(column_data[col_name], bins=20, kde=True)
            
        plt.ylabel('counts')
        plt.xticks(rotation=90)
        plt.title(f'{col_name} (column {i})')
    
    plt.tight_layout(pad=1.0, w_pad=1.0, h_pad=1.0)
    plt.show()

# Assuming df1 is your PySpark DataFrame
plot_per_column_distribution(df_cleaned2, 10, 5)
```

```python
def plot_correlation_matrix(df: DataFrame, graph_width: int):
    # Select only numeric columns
    numeric_cols = [col_name for col_name, data_type in df.dtypes if data_type != 'string']
    
    if len(numeric_cols) < 2:
        print(f'No correlation plots shown: The number of numeric columns ({len(numeric_cols)}) is less than 2')
        return
    
    # Select numeric columns and drop NaNs
    numeric_df = df.select(*numeric_cols).dropna()
    
    # Convert to Pandas DataFrame for correlation computation
    corr = numeric_df.toPandas().corr()
    
    plt.figure(figsize=(graph_width, graph_width), dpi=80, facecolor='w', edgecolor='k')
    corr_mat = sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(f'Correlation Matrix for DataFrame', fontsize=15)
    plt.show()

# Assuming df1 is your PySpark DataFrame
plot_correlation_matrix(df_cleaned2, 10)
```

The independent variable is the cause. Its value is independent of other variables in your study. 

The dependent variable is the effect. Its value depends on changes in the independent variable.  


In traditional statistical modeling and analysis, it's common to have a single dependent variable.

**Dependent Variables:**

**Independent Variables:**

1. **Category:** The category of the app on the Google Play Store.

In this setup, I am interested in understanding how the independent variables (Category, Type, Price, Genres) influence each of the three dependent variables (Rating, Reviews, Installs). 

```python
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline

# String Indexing
category_indexer = StringIndexer(inputCol='Category', outputCol='CategoryIndex')
type_indexer = StringIndexer(inputCol='Type', outputCol='TypeIndex')
genres_indexer = StringIndexer(inputCol='Genres', outputCol='GenresIndex')

# One-Hot Encoding
category_encoder = OneHotEncoder(inputCol='CategoryIndex', outputCol='CategoryVec')
type_encoder = OneHotEncoder(inputCol='TypeIndex', outputCol='TypeVec')
genres_encoder = OneHotEncoder(inputCol='GenresIndex', outputCol='GenresVec')

# Feature Assembling
feature_assembler = VectorAssembler(
    inputCols=['CategoryVec', 'TypeVec', 'Price', 'GenresVec'],
    outputCol='IndependentFeatures'
)

# Model Training
lr = LinearRegression(labelCol="Rating", featuresCol="IndependentFeatures")

# Pipeline
pipeline = Pipeline(stages=[category_indexer, type_indexer, genres_indexer, 
                            category_encoder, type_encoder, genres_encoder, 
                            feature_assembler, lr])

# Fit the pipeline
model = pipeline.fit(training)

# Transform the data
transformed_data = model.transform(training)
```
